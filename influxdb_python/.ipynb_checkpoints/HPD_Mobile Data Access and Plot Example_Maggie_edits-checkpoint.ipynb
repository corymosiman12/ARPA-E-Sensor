```python
import influxdb
import pandas as pd
import matplotlib.pyplot as plt
%pylab inline
plt.rcParams['figure.figsize'] = [16, 9]
```

```python
# Access localhost:8088 and use database test5
c = influxdb.DataFrameClient(host="localhost", port=8088, database="test12")

# Data always stored in env_params table.  Grab all data.
q1 = "select * from env_params"
r = c.query(q1)

# Query returns dict of dataframes, so just grab dataframe from the env_params key
my_data = r['env_params']
```

```python
my_data.info()
```

```python
# Check the column names
my_data.columns
```

```python
my_data.groupby('server_id')['co2eq_ppm'].plot(legend=True, title='CO2 Readings')#, xlim=('2018-12-12', '2018-12-17'))
```

```python
my_data.groupby('server_id')['co2eq_base'].plot(legend=True, title='CO2 Base Readings', xlim=('2018-12-12', '2018-12-17'))
```

```python
my_data.groupby('server_id')['temp_c'].plot(legend=True, title='Temperature Readings', xlim=('2018-12-12', '2018-12-17'))
```

```python
my_data.groupby('server_id')['rh_percent'].plot(legend=True, title='Relative Humidity Readings', ylim=[0,100], xlim=('2018-12-12', '2018-12-17'))
```

```python
my_data.groupby('server_id')['dist_mm'].plot(legend=True, title='Distance Readings', xlim=('2018-12-12', '2018-12-17'))
```

```python
my_data.groupby('server_id')['tvoc_ppb'].plot(legend=True, title='TVOC Readings', xlim=('2018-12-12', '2018-12-17'))
```

```python
my_data.groupby('server_id')['tvoc_base'].plot(legend=True, title='TVOC Base Readings', xlim=('2018-12-12', '2018-12-17'))
```

```python
my_data.groupby('server_id')['light_lux'].plot(legend=True, title='Light Level Readings', xlim=('2018-12-12', '2018-12-17'))
```
